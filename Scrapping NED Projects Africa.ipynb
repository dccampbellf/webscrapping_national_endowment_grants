```python
from bs4 import BeautifulSoup
import requests
import unicodedata
import numpy as np
import pandas as pd
```

```python
responsa = requests.get('https://www.ned.org/region/africa/')
soup_cl=BeautifulSoup(responsa.text,'html.parser')
print(soup_cl.prettify())
```

```python
for link in soup_cl.find_all('option'):
    print(link.text.strip())   
```

```python
table = soup_cl.find_all('option',attrs={'value':True})

country_urls=[]

for thing in table: 
    country_url.append(thing['value'])
    
# using list comprehension to 
# perform removal 

country_urls = [i for i in country_url if i]
country_urls
```

```python
x=3
countries_all_info=[]

for country in country_urls:
    try:
        print(country)
        response = requests.get(country)
        soup=BeautifulSoup(response.text,'html.parser')
        posts = soup.find_all('p')
        
        for post in posts:
            if "1025 F Street NW. Suite 800, Washington DC 20004" in post:
                continue
            else:
                countries_all_info.append(post.text.strip())
                x=x+1
                
                if x%4==0:
                    countries_all_info.append(soup.title.text.strip())
                else:
                     continue
                 
    except:
        continue
```
